# An API explainer notebook

We use this notebook to explain APIs. Let's say we want to download the Rembrandt page https://en.wikipedia.org/wiki/Rembrandt

```python
import urllib.request

url = 'https://da.wikipedia.org/wiki/Stege'
headers = {'User-Agent': 'Mozilla/5.0 (compatible; ExerciseBot/1.0)'} # Set a user-agent to avoid potential blocking by wikipedia
req = urllib.request.Request(url, headers=headers) # Create a request object with the URL and headers
response = urllib.request.urlopen(req)
data = response.read() # a `bytes` object
text = data.decode('utf-8')
```

Now, use the API to get the page in a format we like

https://en.wikipedia.org/w/api.php?action=query&titles=Rembrandt&prop=revisions&rvprop=content&format=json

```python
baseurl = "https://en.wikipedia.org/w/api.php?"
action = "action=query"
title = "titles=Stege"
content = "prop=revisions&rvprop=content"
dataformat ="format=json"

query = "{}{}&{}&{}&{}".format(baseurl, action, content, title, dataformat)
print(query)
```

```python
req_query = urllib.request.Request(query, headers=headers)
wikiresponse = urllib.request.urlopen(req_query)
wikidata = wikiresponse.read()
wikitext = wikidata.decode('utf-8')
```

```python
import json
json_wikitext = json.loads(wikitext)
```
