**Define a function called vocab_size(text) that has a single parameter for the text, and which returns the vocabulary size of the text.**

```python
def vocab_size(text):
    '''
    Returns the vocabulary size of the text.
    But doesn't distinguish between upper and lower case words.
    This is done such that if the word is the first word in a sentence, it is not counted as a different word than the same word in the middle of the sentence.
    '''

    return len(set(text.lower()))
```

**Define a function percent(word, text) that calculates how often a given word occurs in a text, and expresses the result as a percentage.**

```python
def percent(word, text):
    '''
    Returns the percentage of how many times the word appears in the text as a fraction of the length of the text.
    Compares the word in a non case-sensitive manner.
    '''
    return 100 * text.lower().count(word.lower()) / len(text)
```

**4. Read in the texts of the State of the Union addresses, using the state_union corpus reader. Count occurrences of men, women, and people in each document. What has happened to the usage of these words over time?**

```python
from nltk.corpus import state_union

texts = state_union.fileids() 
print(texts[0:10])
```

```python
men_count = []
women_count = []
people_count = []

# count number of occurrences
for fileid in texts:
    words = [w.lower() for w in state_union.words(fileid)]
    # count the number of occurrences of each word
    men_count.append(words.count("men"))
    women_count.append(words.count("women"))
    people_count.append(words.count("people"))

print(men_count)
print(women_count)
print(people_count)

# plot the occurrences over time
years = [int(fileid[:4]) for fileid in texts]
plt.plot(years, men_count, label='Men')
plt.plot(years, women_count, label='Women')
plt.plot(years, people_count, label='People')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Word Occurrences Over Time')
plt.legend()
plt.show()
```

**16. Write a program to generate a table of lexical diversity scores (i.e. token/type ratios), as we saw in 1.1. Include the full set of Brown Corpus genres (nltk.corpus.brown.categories()). Which genre has the lowest diversity (greatest number of tokens per type)? Is this what you would have expected?**

```python

def lexical_diversity(text):
    return len(set(text)) / len(text)

def percentage(count, total):
    return 100 * count / total
```

```python
import nltk
from nltk.corpus import brown

# getting the genres in the brown corpus
genres = brown.categories()

# create a table of lexical diversity scores for each genre
diversity_scores = {}
for genre in genres:
    words = brown.words(categories=genre)
    diversity_scores[genre] = lexical_diversity(words)

# finding the genre with the lowest diversity
lowest_diversity_genre = min(diversity_scores, key=diversity_scores.get)
print(f"Genre with the lowest diversity: {lowest_diversity_genre} with score {diversity_scores[lowest_diversity_genre]:.4f}")
```

```python
# printing the diversity scores based on genre
for genre, score in diversity_scores.items():
    print(f"{genre}: {score:.2f}")
```

**17. ◑ Write a function that finds the 50 most frequently occurring words of a text that are not stopwords.**

```python
def most_frequent_non_stopwords(text, num_words=50):
    from nltk.corpus import stopwords
    from nltk import FreqDist

    # get stopwords from the library
    stop_words = set(stopwords.words('english'))

    # filter out stopwords in the text
    words = [w.lower() for w in text if w.lower() not in stop_words]

    # create frequency distribution to find most common words
    fdist = FreqDist(words)
    most_common = fdist.most_common(num_words)
    
    return most_common

from nltk.book import text1

# Example usage with text5
common_words = most_frequent_non_stopwords(text1, 50)

print("10 most frequent non-stopwords in text1:")
for word, freq in common_words[:10]:  # print only the top 10 for brevity
    print(f"{word}: {freq}")
```

**18. ◑ Write a program to print the 50 most frequent bigrams (pairs of adjacent words) of a text, omitting bigrams that contain stopwords**

```python
def most_frequent_bigrams(text, num_words=50):
    from nltk.corpus import stopwords
    from nltk import FreqDist
    from nltk import bigrams

    # make text lowercase
    text = [w.lower() for w in text]
    
    # get bigrams from the text
    bigram_list = list(bigrams(text))

    # get stopwords from the library
    stop_words = set(stopwords.words('english'))

    # filter out bigrams if they contain any stopwords
    bigram_list = [bg for bg in bigram_list 
                   if bg[0].lower() not in stop_words 
                   and bg[1].lower() not in stop_words]

    # create frequency distribution to find most common bigrams
    fdist = FreqDist(bigram_list)
    most_common = fdist.most_common(num_words)

    return most_common
```

```python
# Example usage with text1
common_bigrams = most_frequent_bigrams(text1, 50)

print("10 most frequent bigrams in text1:")
for bigram, freq in common_bigrams[:15]:  # print only the top 10 for brevity
    print(f"{bigram}: {freq}")
```
