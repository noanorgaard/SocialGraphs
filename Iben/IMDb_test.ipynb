```python

import imdb

# Initialize IMDb instance
ia = imdb.IMDb()

# Get movie by IMDb ID (example: The Matrix = '0133093')
movie = ia.get_movie('0133093')

# update movie to get full details
ia.update(movie)

# Title and year
print("Title:", movie.get('title'))
print("Year:", movie.get('year'))

# Director(s)
directors = movie.get('director')
if directors:
    print("Directors:", [d['name'] for d in directors])

# Cast (first 10 actors/actresses)
cast = movie.get('cast')
if cast:
    print("Cast:", [c['name'] for c in cast[:10]])

# Genres
genres = movie.get('genres')





```

```python
# print the atributes of the movie object
for attribute in movie.keys():
    print(attribute)
    print(movie[attribute])
    print("\n")

# get the director(s) of the movie
```
